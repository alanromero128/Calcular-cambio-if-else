## Las variables
1. CP --> Costo del producto
2. CD --> Cantidad de dinero
3. Cambio  

## Las formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChart

<img src="https://raw.githubusercontent.com/alanromero128/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20cambio2.png" width="300" height="500" />





```python
# para cuando es verdadero (si)
# Calculando cambio usando if else
# Datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if  CP >= CD:
    print("No alcanza")
else :
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Alan Romero NL = 47")
```

```python
# para cuando es falso (no)
# Calculando cambio usando if else
# Datos de entrada CD , CP
# mostrar Cambio si es falso
# si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if  CP >= CD:
    print("No alcanza")
else :
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Alan Romero NL = 47")
```
